```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Sample data with closely spaced peaks
x = [0, 10, 0, 10, 0, 20, 0, 0,0,10,1,16,28,29, 30, 14,6, 10, 0, 20, 0]

# Detect peaks with distance and width criteria
peaks_distance, _ = find_peaks(x, distance=3)
peaks_width, properties = find_peaks(x, width=4)

# Check if any peaks were detected based on width criteria
if len(peaks_width) == 0:
    print("No peaks met the width >= 3 criteria.")

# Plot the data
plt.plot(x, label="Data")
plt.plot(peaks_distance, np.array(x)[peaks_distance], "o", label="Peaks (distance=3)", color="orange")
plt.plot(peaks_width, np.array(x)[peaks_width], "x", label="Peaks (width >= 3)", color="green")

# Highlight width for detected peaks
for i, peak in enumerate(peaks_width):
    left_ip = int(properties["left_ips"][i])  # Start of width
    right_ip = int(properties["right_ips"][i])  # End of width
    plt.hlines(y=x[peak]/2, xmin=left_ip, xmax=right_ip, color="orange", linestyle="--")

plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.title("Distance vs Width in Peak Detection")
plt.show()
```
